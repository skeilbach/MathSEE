```python
import umbridge
import pandas as pd
model = umbridge.HTTPModel("http://localhost:4243", "forward")

R1 = range(0,11)
R2 = range(0,11)

c = 1
d2 = 2
e = 1
f = 2
g = 1
h = 2
i = 1
j = 2
k = 1
l = 2
m = 1
n = 2

d = pd.DataFrame({})
for a in R1:
    for b in R2:
        new_row = pd.DataFrame({'a': a,'b': b,'Temperature':  model([[a,b,c,d2,e,f,g,h,i,j,k,l,m,n]])})
        d = pd.concat([d, new_row], axis=0)

d
```
